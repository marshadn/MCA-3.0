```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs


data,_=make_blobs(n_samples=300)
#print(data)

kmeans=KMeans(n_clusters=5,random_state=42,n_init='auto' ).fit(data)
labels=kmeans.labels_ #it contains the cluster (label) for each data point specifying which cluster is each datapoint belongs to

set(labels)
centroids = kmeans.cluster_centers_ # stores coordinates of centroids for these 5 clusters




plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
#coloring each data points according to the cluster label

plt.scatter(centroids[:, 0], centroids[:, 1], c='red')


plt.legend(list(labels))
kmeans.predict([[2,-4],[6,6],[6,-2]])



```
