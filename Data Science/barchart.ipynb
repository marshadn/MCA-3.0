```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
groups=['A','B','C','D','E']
men=[22,30,35,35,26]
women=[25,32,30,35,29]
```

```python
barwidth=0.35
index=np.arange(len(groups))
```

```python
# print(index)
```

```python
plt.bar(index,men,barwidth,label="Men")
plt.bar(index+barwidth,women,barwidth,label="Women")
plt.xticks(index+barwidth/2,groups)
plt.title("Barchart Groupby Genders")
plt.xlabel("Genders")
plt.ylabel("Mean scores")
plt.legend()
plt.show()
```
