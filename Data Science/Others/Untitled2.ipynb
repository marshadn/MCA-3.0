```python
# #matplotlib

from matplotlib import pyplot as plt
# #OR
# import matplotlib.pyplot as plt

# #plt.show()

import numpy as np
#import pyplot as plt
x=np.arange(1,11)
y=x*x*x+5
#title
plt.title("Title")
plt.xlabel("X axis")
plt.ylabel("Y Axis")
plt.plot(x,y,color='red',marker='o',markersize=12,label='graph')
plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0,3*np.pi,0.1)
y=np.tan(x)
#y=np.sin(x)
#y=np.tan(x)
plt.title("Hello")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,'o',label='graph')
plt.legend()
plt.show()
```

```python
#To include multiple graphs in a single

x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)

plt.subplot(2,1,1)
plt.title("sine")
plt.plot(x,z)
plt.plot(x,y)
plt.show()
```
