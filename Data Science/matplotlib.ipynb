# Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
y = x + 5
```

```python
plt.title("Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.plot(x, y, '--', color='c')
plt.show()
```

```python
y = 3 * x + 5
plt.plot(x, y, color='red', marker='o', markersize='5', label='graph')
plt.legend()
plt.show()
```

```python
x = [2005, 2005, 2015, 2025, 2035]
y = [100, 159, 210, 230, 305]
z = [10, 26, 35, 53, 76]
```

```python
plt.plot(x, y, label="Plot")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.title("Sine Wave")
plt.plot(x, y, color='r', label='sin')
plt.legend()
plt.show()
```

```python
y = np.cos(x)

plt.title("Cosine Wave")
plt.plot(x, y, color='y', label="cos")
plt.legend()
plt.show()
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
plt.subplot(2, 2, 1) # height, width, active
plt.plot(x, y)
plt.title("Sine Wave")

plt.subplot(2, 2, 1)
plt.plot(x, z)
plt.title("Cosine Wave")
plt.show()
```

```python
x = [5, 8, 10]
y = [12, 16, 6]

x2 = [6, 9, 11]
y2 = [6, 15, 7]
```

```python
plt.bar(x, y, align='center')
plt.show()
```

```python
plt.title("Red bars")
plt.bar(x2, y2, color='r')
plt.show()
```

```python
plt.pie(x, colors=['red', 'blue', 'orange'], labels=['A', 'B', 'C'], autopct="%1.1f%%", explode=(0, 0, 0.1), startangle=90)
plt.show()
```

```python
plt.hist(np.random.normal(150, 1, 100), color='lime', edgecolor='green')
print(y)
plt.show()
```

Write a python code to plot a line chart to depict the run rate of T20 Match from the given data: 

Overs | Runs
---|---
5 | 45
10 | 79
15 | 145
20 | 234

Change the thickness of the line.

```python
x = [5, 10, 15, 20]
y = [45, 79, 145, 234]
```

```python
plt.plot(x, y, label='Run Rate', linewidth='5')

plt.title("Run Rate")
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```
